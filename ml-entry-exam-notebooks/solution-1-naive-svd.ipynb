```python
# Aproach 1: Model-based collaborative filtering system
# Baseline: don't consider user/item features at all
# Source: https://www.kaggle.com/shawamar/product-recommendation-system-for-e-commerce
df_train_mod = df_train
df_train_mod['like'] = np.where(df_train_mod['like'] == 1, 1, -1)
utility_mat = df_train_mod.pivot_table(values='like', index='user_id', columns='item_id', fill_value=0)
X = utility_mat.T
print(X.shape)
X.head()
```

```python
from sklearn.decomposition import TruncatedSVD

SVD = TruncatedSVD(n_components=10)
decomposed_matrix = SVD.fit_transform(X)
print(decomposed_matrix.shape)
#decomposed_matrix
```

```python
import numpy as np

correlation_matrix = np.corrcoef(decomposed_matrix)
print(correlation_matrix.shape)
#correlation_matrix
```

```python
item_id = 42
correlation_product_ID = correlation_matrix[item_id]
print(correlation_product_ID.shape)
#correlation_product_ID
```

```python
Recommend = list(X.index[correlation_product_ID > 0.7])
Recommend
```

```python
# {user, [(liked_item_0, time), ... ]}
user_old_preferences_dct = dict()
for index, row in df_train[df_train['like'] == 1].iterrows():
    user_id, item_id, time = row['user_id'], row['item_id'], row['timestamp']
    if user_id not in user_old_preferences_dct:
        user_old_preferences_dct[user_id] = [(item_id, time)]
    else:
        user_old_preferences_dct[user_id].append((item_id, time))
```

```python
def get_top_k_simular_to(items, blacklist, k):
    res = []
    for item_id in items:
        for other_item_id, corr_score in enumerate(correlation_matrix[item_id]):
            if other_item_id not in blacklist:
                res.append((other_item_id, corr_score))
    res_sorted_by_score = sorted(res, key=lambda t: t[1], reverse=True)
    if len(res_sorted_by_score) < k:
        raise
    return res_sorted_by_score[0:k]
```

```python
user_predicted_preferences_dct = dict()
for index, row in df_test.iterrows():
    user_id, timestamp = row['user_id'], row['timestamp']
    all_preferences = [t[0] for t in user_old_preferences_dct[user_id]]
    blacklist = [t[0] for t in user_old_preferences_dct[user_id] if t[1] < timestamp]
    user_predicted_preferences_dct[user_id] = get_top_k_simular_to(all_preferences, blacklist, 20)
#user_predicted_preferences_dct
```

```python
user_predicted_lists = dict()
for user_id, scored_values in user_predicted_preferences_dct.items():
    user_predicted_lists[user_id] = [t[0] for t in scored_values]
#user_predicted_lists
```

```python
df_subb = pd.DataFrame(user_predicted_lists).T
df_subb.index.rename('user_id', inplace=True)
#df_subb.to_csv("baseline.csv")
df_subb.head()
```
